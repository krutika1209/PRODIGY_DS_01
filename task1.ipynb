# distribution of age according to city 

```python
age_distribution = df[['name_of_city', '0-6_population_total', '0-6_population_male', '0-6_population_female']]
```

```python
age_distribution_melted = age_distribution.melt(id_vars=['name_of_city'], 
                                                value_vars=['0-6_population_total', '0-6_population_male', '0-6_population_female'], 
                                                var_name='Age_Group', 
                                                value_name='Population')
```

```python
plt.figure(figsize=(14, 8))
sns.barplot(data=age_distribution_melted, x='name_of_city', y='Population', hue='Age_Group')
plt.title('Age Distribution (0-6 years) by City')
plt.xlabel('City')
plt.ylabel('Population')
plt.xticks(rotation=90)
plt.legend(title='Age Group')
plt.show()
```

```python
# List of numerical columns for univariate analysis
numerical_columns = [
    'population_total', 'population_male', 'population_female', 
    '0-6_population_total', '0-6_population_male', '0-6_population_female', 
    'literates_total', 'literates_male', 'literates_female', 
    'sex_ratio', 'child_sex_ratio', 
    'effective_literacy_rate_total', 'effective_literacy_rate_male', 'effective_literacy_rate_female', 
    'total_graduates', 'male_graduates', 'female_graduates'
]

# Univariate analysis for numerical columns
for column in numerical_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[column], kde=True, bins=10)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# List of pair columns for bivariate analysis
pair_columns = [
    ('population_total', 'literates_total'),
    ('population_total', 'total_graduates'),
    ('population_male', 'population_female'),
    ('literates_male', 'literates_female'),
    ('effective_literacy_rate_male', 'effective_literacy_rate_female'),
    ('total_graduates', 'male_graduates'),
    ('total_graduates', 'female_graduates')
]

# Bivariate analysis using scatter plots
for x_col, y_col in pair_columns:
    plt.figure(figsize=(20, 10))
    sns.barplot(x=df[x_col], y=df[y_col])
    plt.title(f'Bar Plot of {x_col} vs {y_col}')
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.show()


```
